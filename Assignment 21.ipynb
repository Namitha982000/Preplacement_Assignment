#### Question-1 You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def inOrderTraversal(root, values):
    if root is None:
        return
    
   
    inOrderTraversal(root.left, values)
    values.append(root.val)
    inOrderTraversal(root.right, values)

def constructBST(root, values):
    if root is None:
        return
    
   
    constructBST(root.left, values)
    root.val = values.pop(0)
    constructBST(root.right, values)

def convertToBST(root):
    values = []
    inOrderTraversal(root, values)
    values.sort()
    constructBST(root, values)


def printInOrder(root):
    if root is None:
        return
    
    printInOrder(root.left)
    print(root.val, end=" ")
    printInOrder(root.right)


root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

print("Input Binary Tree (before conversion to BST):")
printInOrder(root)

convertToBST(root)

print("\nBinary Search Tree (after conversion):")
printInOrder(root)
```

#### Question-3: Write a program to convert a binary tree to a doubly linked list.


```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

def inorderTraversal(root):
    if root is None:
        return []
    return inorderTraversal(root.left) + [root] + inorderTraversal(root.right)

def convertToDoublyLinkedList(root):
    nodes = inorderTraversal(root)
    dummy = DoublyLinkedListNode(None)
    last = dummy

    for node in nodes:
        new_node = DoublyLinkedListNode(node.val)
        new_node.prev = last
        last.next = new_node
        last = new_node

    return dummy.next

root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

head = convertToDoublyLinkedList(root)

print("\nOutput Doubly Linked List:")
while head is not None:
    print(head.val, end=" ")
    head = head.next
```
